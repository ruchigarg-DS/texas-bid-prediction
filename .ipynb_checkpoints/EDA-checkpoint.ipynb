```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('max_colwidth', 800)  # displays complete text
pd.set_option('float_format', '{:.2f}'.format) # displays float with 2 decimal spaces

# Read the excel sheet using pandas
bidsData = pd.read_csv('Dataset/texas-bids.csv')

# display the 1st 2 rows
bidsData.head(2)
```

```python
# Check the shape of the dataset
bidsData.shape
```

```python
# Get column typed for dataset
bidsData.info()
```

```python
bidsData.describe()
```

```python
# Remove the non-numerical characters and convert the time string into int
bidsData['time'] = [int(''.join(c for c in each if c.isdigit())) for each in bidsData['time']] 

#  Convert tim column to type int
bidsData['time'].astype(int)
bidsData.info()
```

```python
bidsData.to_csv('Dataset/cleaned_bids.csv', index=False)
```

```python
# Splitting the compiled_bids column into a list
# bidsList = [each.split(';') for each in bidsData['compiled_bids']]
```

```python
bidsData['compiled_bids']=bidsData['compiled_bids'].replace('\$','',regex=True)
```

```python
testBids = bidsData['compiled_bids'][0]
testBids
```

```python
bidsList = [[each.replace('|', ':') for each in value.split(';')] for value in bidsData['compiled_bids']]
```

```python
# [[each.split(',') for each in value] for value in bidsList]#each.split(',') for each in bidsList[0]]

# bidsVals = [each.split(',') for each in bidsList[0]]

# bidsVals

bidsSeries = pd.Series(bidsList)
bidsSeries

bidsData['bids'] = bidsSeries.values
```

```python
# Creating a dictionary using the proj_id as key and bidsList as values
bids = dict(zip(bidsData['proj_ID'], bidsList))
dict(list(bids.items())[0:2])
```
