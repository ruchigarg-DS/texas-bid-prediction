```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
sns.set()


# pd.set_option('max_colwidth', 800)  # dispalys complete text
pd.set_option('float_format', '{:.2f}'.format) # displays float with 2 decimal spaces
```

```python
# Read the excel sheet using pandas
bidsData = pd.read_csv('Dataset/cleaned_bids.csv')

# display the 1st 2 rows
bidsData.head(2)
```

```python
oldcastleDataset = bidsData[bidsData['winner'] == 'OLDCASTLE MATERIALS TEXAS INC.']
```

```python
sns.heatmap(oldcastleDataset.corr(), annot=True)
```

```python
sns.pairplot(oldcastleDataset)
```

```python
sns.jointplot(x="estimate",y="win_amt", data=oldcastleDataset, kind='scatter')
```

```python
sns.jointplot(y="estimate",x="num_bids", data=oldcastleDataset, kind='reg')
```

```python
sns.jointplot(x="estimate",y="time", data=oldcastleDataset, kind='scatter')
```

```python
sns.jointplot(x="estimate",y="bid_spread", data=oldcastleDataset)
```

```python
g = sns.PairGrid(oldcastleDataset)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
plt.show();
```
