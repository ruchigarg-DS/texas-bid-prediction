```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('max_colwidth', 800)  # displays complete text
pd.set_option('float_format', '{:.2f}'.format) # displays float with 2 decimal spaces
```

```python
# Read the excel sheet using pandas
bidsData = pd.read_csv('Dataset/cleaned_bids.csv')

# display the 1st 2 rows
bidsData.head(2)
```

```python
bigcreekDataset = bidsData[bidsData['winner'] == 'BIG CREEK CONSTRUCTION LTD.']
bigcreekDataset.head()
```

```python
sns.heatmap(bigcreekDataset.corr(), annot=True)
```

```python
sns.jointplot(x="estimate",y="win_amt", data=bigcreekDataset, kind='scatter')
```

```python
sns.jointplot(y="estimate",x="num_bids", data=bigcreekDataset, kind='reg')
```

```python
sns.jointplot(x="time",y="estimate", data=bigcreekDataset, kind='scatter')
```

```python
sns.jointplot(x="estimate",y="bid_spread", data=bigcreekDataset)
```

```python
g = sns.PairGrid(bigcreekDataset)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
plt.show();
```
